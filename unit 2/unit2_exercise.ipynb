```python
'''
4. Plot this exampleâ€™s waveform...
'''
#%% Waveform
import librosa
import librosa.display
import matplotlib.pyplot as plt

example = sublist[-1]

array = example["audio"]["array"]
sampling_rate = example["audio"]["sampling_rate"]

plt.figure().set_figwidth(12)
librosa.display.waveshow(array, sr=sampling_rate)
```

```python
'''
4. ... and spectrogram.
'''
#%% Spectrogram
import numpy as np
D = librosa.stft(array)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

plt.figure().set_figwidth(12)
librosa.display.specshow(S_db, x_axis="time", y_axis="hz")
plt.colorbar()
```

```python
'''
5. Go to ðŸ¤— Hub, explore pretrained models and find one that can be used
for automatic speech recognition for the language that you have picked earlier.
Instantiate a corresponding pipeline with the model you found, and transcribe the example.
'''
#!pip install kenlm pyctcdecode
from transformers import pipeline
pipe = pipeline(model="jonatasgrosman/wav2vec2-large-xlsr-53-polish")
test_transcript = pipe(example["audio"]["array"])

#from transformers import pipeline
#pipe = pipeline("text-to-speech", model="suno/bark-small")
```

```python
'''
6. Compare the transcription that you get from the pipeline to the transcription provided in the example.
'''
!pip install jiwer
from jiwer import wer
error = wer(example['normalized_text'], test_transcript['text'])
print('reference:', example['normalized_text'], '\n',
      'hypothesis:', test_transcript['text'], '\n'
      'error:', error)
```
