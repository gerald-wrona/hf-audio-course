```python
# Are waveform dimensions consistent across all records? 
from collections import Counter

array_dims = []
for record in gtzan['train']:
    array_dims.append(record['audio']['array'].shape[0])

dims = dict(Counter(array_dims))
print('We have', len(dims.keys()), 'unique dimensions counts.',
      '\n', '\n', 'Count of unique dimensionsal patterns:',
      '\n', dims,
     '\n', '\n', 'We see 64% are 661794 and 30% are 661504.')

#print(dims)
```

```python
# As a histogram
import matplotlib.pyplot as plt
plt.hist(array_dims)
plt.show() 
```

A substantial difference would be figures less than or equal to 639,450 (29 seconds) or greater than or equal to 683,550 (31 seconds). Two records come close to 683,550.

```python
# Is the sampling rate consistent across all recordings?
from collections import Counter

s_rates = []
for record in gtzan['train']:
    s_rates.append(record['audio']['sampling_rate'])

s_rate_counts = dict(Counter(s_rates))
print('We have', len(s_rate_counts.keys()), 'unique sampling rate.',
      '\n', '\n', 'Count of unique sampling rates:',
      '\n', s_rate_counts)
```

```python
# What is the distribution by genre? Which classes are best represented?
id2label_fn = gtzan["train"].features["genre"].int2str

genres = []
for record in gtzan['train']:
    genres.append(id2label_fn(record['genre']))

plt.hist(genres)
plt.show() 
```

There is no class imbalance. This doesn't mean the model won't struggle to distinguish similar genres. To me, similarity means having similar rhythmic structure and musical surface as described by the original paper (https://ismir2001.ismir.net/pdf/tzanetakis.pdf).

Pop, reggae, blues and metal are not included in the original paper. They must be later additions. I wasn't sure if all records were instrumentation-only, but the paper suggests they include vocalization. It's another instrument, but if the model is able to distinguish instruments (pre-2001 research was able to) then distinguishing hip hop and classical should be easy; they use entirely different instrument groups and classical often has no vocals. 

In Tzanetakis et al the model accuracy for jazz, rock is less than 50%. They give an explanation that the "boundaries" for these genres are "fuzzier" than others. I think "rock" is a bit broad. This wouldn't be an issue if they were categorizing jazz vs metal, a subgenre of rock. I also think the number of genres and our understanding of genres has increased since 2001. 

I want to test a hypothesis that the model used in the example was performing poorly on certain genres because of their similarity. Classical will always be quite distinct from hiphop because they use different instruments due to being pioneered at vastly different times in history. Blues and jazz share elements and pop tends to borrow from all genres, so I hypothesize their attention-based model struggles when it has found important features that could be applied to both labels. https://huggingface.co/learn/audio-course/chapter4/fine-tuning
